# Titanic: Machine Learning from Disaster

The sinking of the RMS Titanic is one of the most infamous shipwrecks in history.  On April 15, 1912, during her maiden voyage, the Titanic sank after colliding with an iceberg, killing 1502 out of 2224 passengers and crew. This sensational tragedy shocked the international community and led to better safety regulations for ships.

One of the reasons that the shipwreck led to such loss of life was that there were not enough lifeboats for the passengers and crew. Although there was some element of luck involved in surviving the sinking, some groups of people were more likely to survive than others, such as women, children, and the upper-class.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("train.csv")
df2 = pd.read_csv("test.csv")
df3 = pd.read_csv("gender_submission.csv")
```

```python
survived = df.loc[df['Survived'] == 1]
```

```python
unsurvived = df.loc[df['Survived'] == 0]
```

```python
survived_classes_count = survived.groupby('Pclass')['Name'].count()
survived_classes_names = survived['Pclass'].unique()
survived_classes_names = sorted(survived_classes_names)
survived_df = pd.DataFrame({"names":survived_classes_names, "count": survived_classes_count})
fig, ax = plt.subplots(figsize=(6,3))
sb.barplot(ax=ax, x=survived_df['names'], y=survived_df['count'], data=survived_df)
plt.title("The Survival classes comparison");
```

```python
survived_gender_count = survived.groupby('Sex')['Name'].count()
survived_gender_names = survived['Sex'].unique()
survived_gender_names = sorted(survived_gender_names)
survived_gender_df = pd.DataFrame({"names":survived_gender_names, "count": survived_gender_count})
fig, ax = plt.subplots(figsize=(6,3))
sb.barplot(ax=ax, x=survived_gender_df['names'], y=survived_gender_df['count'], data=survived_gender_df)
plt.title("The Survival gender comparison");
```

```python
survived_age_count = survived.groupby('Age')['Name'].count()
survived_age_names = survived['Age'].dropna().unique()
survived_age_names = sorted(survived_age_names)
survived_age_df = pd.DataFrame({"names":survived_age_names, "count": survived_age_count})
fig, ax = plt.subplots(figsize=(36,3))
sb.barplot(ax=ax, x=survived_age_df['names'], y=survived_age_df['count'], data=survived_age_df)
plt.title("The Survival age comparison");
plt.xticks(rotation=-90);
```

```python
unsurvived_gender_count = unsurvived.groupby('Sex')['Name'].count()
unsurvived_gender_names = unsurvived['Sex'].unique()
unsurvived_gender_names = sorted(unsurvived_gender_names)
unsurvived_gender_df = pd.DataFrame({"names":unsurvived_gender_names, "count": unsurvived_gender_count})
fig, ax = plt.subplots(figsize=(6,3))
sb.barplot(ax=ax, x=unsurvived_gender_df['names'], y=unsurvived_gender_df['count'], data=unsurvived_gender_df)
plt.title("The unSurvival gender comparison");
```

```python
unsurvived_classes_count = unsurvived.groupby('Pclass')['Name'].count()
unsurvived_classes_names = unsurvived['Pclass'].unique()
unsurvived_classes_names = sorted(unsurvived_classes_names)
unsurvived_df = pd.DataFrame({"names":unsurvived_classes_names, "count": unsurvived_classes_count})
fig, ax = plt.subplots(figsize=(6,3))
sb.barplot(ax=ax, x=unsurvived_df['names'], y=unsurvived_df['count'], data=unsurvived_df)
plt.title("The unSurvival classes comparison");
```

```python
unsurvived_age_count = unsurvived.groupby('Age')['Name'].count()
unsurvived_age_names = unsurvived['Age'].dropna().unique()
unsurvived_age_names = sorted(unsurvived_age_names)
unsurvived_age_df = pd.DataFrame({"names":unsurvived_age_names, "count": unsurvived_age_count})
fig, ax = plt.subplots(figsize=(36,3))
sb.barplot(ax=ax, x=unsurvived_age_df['names'], y=unsurvived_age_df['count'], data=unsurvived_age_df)
plt.title("The unSurvival age comparison");
plt.xticks(rotation=-90);
```

```python
survived_has_cabin_count = df.loc[(df['Cabin'].notna()) & (df['Survived'] == 1)].count()
survived_has_not_cabin_count = df.loc[(df['Cabin'].isna()) & (df['Survived'] == 1)].count()
unsurvived_has_cabin_count = df.loc[(df['Cabin'].notna()) & (df['Survived'] == 0)].count()
unsurvived_has_not_cabin_count = df.loc[(df['Cabin'].isna()) & (df['Survived'] == 0)].count()
```

```python
labels = "Survived", "UnSurvived"
values = survived_has_cabin_count[0], unsurvived_has_cabin_count[0]
fig, ax = plt.subplots(figsize=(6, 6))
plt.pie(x=values, labels=labels, autopct='%1.2f%%', shadow=True);
plt.title("Passengers That Has Cabin");
```

```python
labels = "Survived", "UnSurvived"
values = survived_has_not_cabin_count[0], unsurvived_has_not_cabin_count[0]
fig, ax = plt.subplots(figsize=(6, 6))
plt.pie(x=values, labels=labels, autopct='%1.2f%%', shadow=True);
plt.title("Passengers That Has No Cabin");
```
